```python
import os
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.tree import export_graphviz 
from IPython.display import Image  
from sklearn import preprocessing

```

```python
col_names = ['outlook', 'temp', 'hum', 'wind', 'playtennis']
# load dataset
pima = pd.read_csv("weather.csv", header=None, names=col_names)
pima.head(10)

pima['playtennis'],class_names = pd.factorize(pima['playtennis'])
pima['outlook'],_ = pd.factorize(pima['outlook'] )
pima['temp'],_ = pd.factorize(pima['temp'] )
pima['hum'],_ = pd.factorize(pima['hum'] )
pima['wind'],_ = pd.factorize(pima['wind'] )
pima['playtennis'],_ = pd.factorize(pima['playtennis'] )
pima.info()
X = pima.iloc[:,:-1]
y = pima.iloc[:,-1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0) # 70% training and 30% test
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
#Predict the response for test dataset
y_pred = clf.predict(X_test)
```
